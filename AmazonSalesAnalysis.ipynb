```python
# Ruta al archivo en tu Drive. Asegúrate de que el nombre y ruta coincidan.
file_path = '/content/drive/MyDrive/AmazonSalesAnalysis/amazon.csv'

# Cargar el CSV en un DataFrame
df = pd.read_csv(file_path)

# Let's check the first few rows to understand the structure of the data
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
# Conocer los datos de mi dataset
```

```python
df['actual_price'] = df['actual_price'].replace('₹', '', regex=True).replace(',', '', regex=True).astype(float)
df['discounted_price'] = df['discounted_price'].replace('₹', '', regex=True).replace(',', '', regex=True).astype(float)
# We will first clean the 'actual_price' and 'discounted_price'column to remove any symbols and convert it to a numeric format
df.head()
```

```python
# Convertir rating a float, forzar errores a NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df.head()
```

```python
df.dropna(subset=['actual_price', 'discounted_price', 'rating'], inplace=True)
# Eliminar filas con datos faltantes
```

```python
df.info()
df.describe()
# Verificar que todo este bien
```

```python
# QUESTION 1:
# What is the distribution of product prices (actual price) in the dataset?
```

```python
# Let's visualize the distribution of 'actual_price' using a histogram
plt.figure(figsize=(10, 6))  # Set the figure size for better readability
sns.histplot(df['actual_price'], kde=True, bins=30, color='blue')  # kde=True adds a kernel density estimate line
plt.title('Distribution of Product Actual Prices')  # Adding title
plt.xlabel('Price (in currency)')  # Label for the x-axis
plt.ylabel('Frequency')  # Label for the y-axis
plt.show()

# Get some basic statistics about 'actual_price' (mean, median, etc.)
print(df['actual_price'].describe())

# We can also check if there are any extreme outliers in 'actual_price' using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['actual_price'], color='orange')
plt.title('Boxplot of Product Actual Prices')
plt.xlabel('Price (in currency)')
plt.show()
```

Key Observations and Conclusions

1- Most Products Are Relatively Affordable

The median price (50%) is 1650, meaning that half of the products cost less than ₹1650.

75% of the products are priced below ₹4303.75.

This shows that the majority of items are in the low to mid-price range.

2- There Are a Few Very Expensive Outliers

The maximum price is ₹139,900, which is much higher than the average (mean) of ₹5447.

The high standard deviation (₹10,878) suggests a wide spread in the prices, likely caused by a small number of very expensive products.
These outliers skew the average, making it higher than the median.

3- The Data is Right-Skewed

Since the mean (₹5447) is greater than the median (₹1650), the distribution of prices is right-skewed (i.e., the tail is longer on the right).
This confirms what you observed in the histogram — most products are clustered on the left (under ₹20,000), and only a few are expensive.

4- Minimum Price is Very Low
The cheapest product costs ₹39, showing that there are very low-budget options in the dataset.

Final Response for Question 1:

Most Amazon products in this dataset are affordable, with a large portion priced under ₹5000.
However, the presence of a few high-end products (up to ₹139,900) creates a long tail in the price distribution.
This impacts the average price, making it higher than the typical product cost.

```python
# QUESTION 2:
# Which product categories have the highest average rating?
```

```python
# Group the dataset by 'category' and calculate the average rating for each category
avg_rating_by_category = df.groupby('category')['rating'].mean().sort_values(ascending=False)

# Display the average ratings for all categories
print("Average Rating by Product Category:")
print(avg_rating_by_category)

# Select the top 10 categories by average rating
top_categories = avg_rating_by_category.head(10)

# Plot the top 10 categories by average rating
plt.figure(figsize=(12, 6))  # Set figure size for better readability

# Create a horizontal barplot using a custom color palette
sns.barplot(
    x=top_categories.values,  # Average ratings
    y=top_categories.index,   # Category names
    palette=sns.color_palette("viridis", len(top_categories))  # Apply 'viridis' palette with matching number of bars
)

# Add chart title and axis labels
plt.title('Top 10 Product Categories by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Product Category')

# Set x-axis limit since ratings typically range from 0 to 5
plt.xlim(0, 5)

# Improve layout and display the plot
plt.tight_layout()
plt.show()
```

Insights from the Average Product Rating by Category

1- Top-Rated Categories:

Products like Tablets, PowerLAN Adapters, Photography Film, Memory Components, and Streaming Devices have average ratings between 4.5 and 4.6, indicating high customer satisfaction. These product types likely offer good performance, reliability, or ease of use that resonates well with users.

2- Lowest-Rated Categories:

At the bottom of the list, we see categories such as:

Electric Coffee Grinders (3.3),

Keyboard & Mouse Dust Covers (3.4),

PC Headsets and 3D Glasses (3.5).
These categories might be suffering from issues like poor design, quality concerns, or unmet user expectations.

3- Rating Range:
The average ratings across all categories range from 3.3 to 4.6, showing a generally positive trend in customer feedback. Most categories seem to maintain a reasonable level of satisfaction.

4- Business Opportunities:

Categories with lower ratings could benefit from a deeper analysis to identify and fix pain points in product design, quality, or customer support.

Meanwhile, top-rated categories present opportunities for marketing, upselling, or even exploring related products with similar traits.

Final response for Question 2:

The analysis reveals that while most Amazon product categories are well received, certain types of products consistently outperform others in terms of customer satisfaction. This insight can help businesses focus their efforts on improving underperforming categories and leverage high-performing ones for strategic growth.
